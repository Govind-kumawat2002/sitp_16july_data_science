```python
# pip install seaborn 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data =sns.load_dataset('tips') ### load dataset from seaborn 
```

```python
data 
```

```python
plt.title("made by matplotlib ")
plt.plot(data['total_bill'])

plt.show()
```

```python
plt.title("made by seaborn ")
sns.lineplot(data['total_bill'])
plt.show()
```

```python
plt.title("This is scatter plot ")
sns.scatterplot(x = 'total_bill', y = 'tip', data= data )
plt.show()
```

```python
plt.title("made by seaborn ")
sns.histplot(data['total_bill'])
plt.show()
```

```python
plt.title("made by matplotlib ")
plt.hist(data['total_bill'])
plt.show()
```

```python
plt.title("This is a kdeplot")
sns.kdeplot(data['total_bill'])
plt.show()
```

```python
# catplot 
# relplot 
```

```python
# plt.title("this is relplot ")
sns.relplot(x = 'total_bill', y = 'tip', data=data)
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=data)
plt.title("total_bill vs tip")
plt.xlabel("total_bill",fontsize = 20, color = 'red')
plt.ylabel("tip",fontsize = 20, color = 'red')
plt.show()
```

```python
#####  catplot 
sns.catplot(x = 'day',  y = 'total_bill', data=data,kind='swarm')
plt.show()
```
