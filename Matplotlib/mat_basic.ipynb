Matplotlib is a data visualization library in Python. It is used to create static, animated, and interactive plots. It is widely used in data science, machine learning, and general Python programming for visualizing data.

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
# plot 
x = np.array([0,6])
y = np.array([0, 250])
plt.plot(x , y ) 
```

```python
# plot 
x = np.array([0,6])
y = np.array([0, 250])
plt.plot(x , y , 'o') ##  
```

```python
# plot 
x = np.array([0,6])
y = np.array([0, 250])
plt.plot(x , y , marker = 'o') ##  
plt.show()
```

```python
y = np.array([0, 2,5,9])
plt.plot( y , marker = 'o') ##  
plt.show()
```

```python
y = np.array([0, 2,5,9])
plt.plot( y , marker = '*') ##  
plt.show()
```

```python
y = np.array([0, 2,5,9])
plt.plot( y , 'o:b') ##  
plt.show()
```

```python
y = np.array([0, 2,5,9])
plt.plot( y , 'o:b', ms = 30) ##  
plt.show()
```

```python
y = np.array([1,5,8,6])
plt.plot(y , marker = "*", ms = 20, mec = "b")
plt.show()
```

```python
y = np.array([1,5,8,6])
plt.plot(y , marker = '*', mec = 'b',ms = 20, mfc = "pink")
plt.show()
```

```python
y = np.array([1,5,8,6])
plt.plot(y , ls='dotted')
plt.show()
```

```python
y = np.array([1,5,8,6])
plt.plot(y , ls='dashed')
plt.show()
```

```python
y = np.array([1,5,8,6])
plt.plot(y , linewidth = 5)
plt.show()
```

```python
x = np.array([1,3,6,8])
y = np.array([2,4,8,7])
plt.plot(x )
plt.plot(y)
plt.show()
```
